## Q1. Given a list of numbers, remove all duplicates and keep the order intact.

```python
# without inbuilt function

def unique_numbers(num_list):
    clean_list=[]
    for i in num_list:
        if i in clean_list:
            pass
        else:
            clean_list.append(i)
    return clean_list

def sort_numbers(numlist):
    for i in range(0,len(numlist)):
        for j in range (i+1,len(numlist)):
            if numlist[i]>numlist[j]:
                temp=numlist[i]
                numlist[i]=numlist[j]
                numlist[j]=temp
    return numlist

a=[1,4,4,4,2,2,6,5]

sort_numbers(unique_numbers(a))
```

```python
#with inbuilt function

a = [1,4,4,4,2,2,6,5]
unique_sorted = sorted(set(a), reverse=False)
print(unique_sorted)
```

**Quick insights:**
- list.sort(): In-place sorting of a list, **modifies the original list**, returns None.
- sorted(): Built-in function, **returns a new sorted list (or any iterable), leaves original unchanged**.
- DataFrame.sort_values(): Pandas method, sorts rows of a DataFrame/Series by column values, more powerful than the deprecated .sort().


## Q2.Write a function to find the second largest number in a list.

```python
lst = [10,40,20,30,50]

def second_largest(num):
    new_list =sorted(set(num))
    new_list.pop()
    return max(new_list)
second_largest(lst)
```

‚ö†Ô∏è Edge cases
- If the list has fewer than 2 unique elements, this will fail (max([]) ‚Üí ValueError).
- Example: [5,5,5] ‚Üí after set ‚Üí [5], .pop() leaves empty list, max([]) errors.

```python
def second_largest(num):
    new_list = sorted(set(num))
    if len(new_list) < 2:
        return None
    return new_list[-2]

print(second_largest([10,40,20,30,50]))  # 40
print(second_largest([5,5,5]))           # None
print(second_largest([10,20]))           # 10
```

```python
def reverse_list(lst):
    r_list=[]
    for i in range(len(lst)-1,-1,-1):
        r_list.append(lst[i])
    return r_list
```

```python
reverse_list(lst)
```

#### with recursion

```python
def rev_list(lst):
    if len (lst)==0:
        return []
    return [lst[-1]]+rev_list(lst[:-1])
```

```python
rev_list(lst)
```

- Base case: If the list has 0 or 1 element, just return it.
- Recursive step: Take the last element (lst[-1]) and put it in front of the reversed version of the rest (lst[:-1]).
- This keeps peeling off the last element until the list is empty


## Q4.Merge two lists and remove common elements.

```python
lst1 = [1,3,5,7]
lst2 = [1,3,4,6]

def merge_list(list1,list2):
    l1=[]
    l2=[]
    for i in list1:
        if i not in list2:
            l1.append(i)
    for i in list2:
        if i not in list1:
            l2.append(i)
    return l1+l2

merge_list(lst1,lst2)
```

```python
#Simplified Version (using set symmetric difference)
#You can achieve the same result more concisely with sets:

def merge_list(list1, list2):
    return list(set(list1) ^ set(list2))

print(merge_list(lst1, lst2))  # [4, 5, 6, 7]

#  ^ is the symmetric difference operator: it returns elements in either set but not both.
```

In Python sets:
- & (ampersand) ‚Üí intersection ‚Üí elements common to both sets.
- ^ (caret) ‚Üí symmetric difference ‚Üí elements in either set, but not in both

## Q11. Count the frequency of each word in a sentence and store it in a dictionary.

```python
sent = ["Count the frequency of each word in a sentence and store it in a dictionary"]

words = ' '.join(sent).lower().split()

res={i:words.count(i) for i in words}

print(res,end='')
```

{i: sent.count(i) for i in sent}
works, but it‚Äôs inefficient because:
- sent.count(i) scans the entire list for each word.
- That makes it O(n¬≤) time complexity.


```python
from collections import Counter

sent = ["Count the frequency of each word in a sentence and store it in a dictionary"]
words = ' '.join(sent).lower().split()
res = dict(Counter(words))
print(res)
```

## Q16. Print the Fibonacci sequence up to n terms using a loop.

```python
def fibonacci(num):
    a=0
    b=1
    for i in range(num):
        print(a,end='')
        a,b=b,a+b
fibonacci(10)
```

## Q17. Write a loop to check if a number is prime.

```python
def is_prime(num):
    for i in range(2,int(num**0.5)+1):
        if num%i == 0:
            return "Not a Prime number"
    return "Prime Number"

is_prime(29)
```

## Q19. Given a list of numbers, use a loop to separate even and odd numbers into two lists.

```python
even = []
odd = []

[even.append(i) if i%2==0 else odd.append(i) for i in range(1,20)]

print(even)
print(odd)
```

‚ö†Ô∏è Note
Using list comprehensions for side effects (like .append()) is considered unpythonic. A cleaner way is:


```python
even, odd = [], []

for i in range(1, 20):
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)

print(even)
print(odd)
```

## Q20. Write a loop to calculate the factorial of a number without recursion.

```python
def factorial(num):
    res = 1
    for i in range(1,num+1):
        res *= i
    return res

factorial(int(input("enter")))
```

with recursion

```python
def factorial(num):
    if num == 0 or num == 1:
        return 1
    return num * factorial(num-1)
print(factorial(int(input("enter: "))))
```

üöÄ Challenge Combo
- Take a list of student records (name, marks) stored as tuples.
- Convert it into a dictionary with names as keys and marks as values.
- Use a loop to print "Pass" if marks ‚â• 40, else "Fail".
- Bonus: Sort the dictionary by marks and print the top scorer.


```python
students = [
    ("Ashwin", 85),
    ("Raj", 92),
    ("Meena", 78),
    ("Kumar", 95),
    ("Divya", 88),
    ("Arun", 91),
    ("Ravi", 92)
]
```

```python
students_dict={i:j for i , j in students}

for key,value in students_dict.items():
    if value >= 40:
        print(key , "Pass")
    else:
        print(key , "Fail")
        
unique_marks = sorted(set({value for key,value in students_dict.items()}),reverse=True)[:3]

unique_marks

top_scorers={key:value for key, value in sorted(students_dict.items(),key = lambda x:x[1],reverse=True) if value in unique_marks}

top_scorers
```

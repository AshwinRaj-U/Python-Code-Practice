## Q1. Given a list of numbers, remove all duplicates and keep the order intact.

```python
# without inbuilt function

def unique_numbers(num_list):
    clean_list=[]
    for i in num_list:
        if i in clean_list:
            pass
        else:
            clean_list.append(i)
    return clean_list

def sort_numbers(numlist):
    for i in range(0,len(numlist)):
        for j in range (i+1,len(numlist)):
            if numlist[i]>numlist[j]:
                temp=numlist[i]
                numlist[i]=numlist[j]
                numlist[j]=temp
    return numlist

a=[1,4,4,4,2,2,6,5]

sort_numbers(unique_numbers(a))
```

## Q2.Write a function to find the second largest number in a list.

```python
lst = [10,40,20,30,50]

def second_largest(num):
    new_list =sorted(set(num))
    new_list.pop()
    return max(new_list)
second_largest(lst)
```

‚ö†Ô∏è Edge cases
- If the list has fewer than 2 unique elements, this will fail (max([]) ‚Üí ValueError).
- Example: [5,5,5] ‚Üí after set ‚Üí [5], .pop() leaves empty list, max([]) errors.

```python
def second_largest(num):
    new_list = sorted(set(num))
    if len(new_list) < 2:
        return None
    return new_list[-2]

print(second_largest([10,40,20,30,50]))  # 40
print(second_largest([5,5,5]))           # None
print(second_largest([10,20]))           # 10
```

## Q4.Merge two lists and remove common elements.

```python
lst1 = [1,3,5,7]
lst2 = [1,3,4,6]

def merge_list(list1,list2):
    l1=[]
    l2=[]
    for i in list1:
        if i not in list2:
            l1.append(i)
    for i in list2:
        if i not in list1:
            l2.append(i)
    return l1+l2

merge_list(lst1,lst2)
```

```python
#Simplified Version (using set symmetric difference)
#You can achieve the same result more concisely with sets:

def merge_list(list1, list2):
    return list(set(list1) ^ set(list2))

print(merge_list(lst1, lst2))  # [4, 5, 6, 7]

#  ^ is the symmetric difference operator: it returns elements in either set but not both.
```

In Python sets:
- & (ampersand) ‚Üí intersection ‚Üí elements common to both sets.
- ^ (caret) ‚Üí symmetric difference ‚Üí elements in either set, but not in both

```python
# My Initial Approach

import string

#--------------Common an quick way to cleanpunctiations
translator=str.maketrans('','',string.punctuation)

clean = [i.translate(translator).lower() for i in sentences]

clean
#---------------
words=[]
for i in clean:
    words.extend(i.split())

from collections import Counter

sorted(set([(i,words.count(i))for i in words]),key = lambda x:(-x[1],x[0]))[:3]
```

```python
from collections import Counter
import string

translator = str.maketrans('', '', string.punctuation)
clean = [s.translate(translator).lower() for s in sentences]

words = []
for s in clean:
    words.extend(s.split())

counter = Counter(words)
print(counter.most_common(3))

# - most_common(3) ‚Üí returns the top 3 items sorted by frequency (highest first).
#- If two items have the same frequency, their order depends on when they first appeared in the data (unless you sort explicitly).
```

```python
# Final Learning

from collections import Counter
import string
def freq_count(data,n):
    translator=str.maketrans('','',string.punctuation)
    clean = [i.translate(translator).lower() for i in data]
    words=[]
    for i in clean:
        words.extend(i.split())   

    counter = Counter(words).most_common() # used for sorting in descending order
    #result = sorted(counter.items(),key = lambda x:(-x[1],x[0])) used for manual sort
    
    return counter[:n]
    
freq_count(sentences,3)
```

The maketrans() method returns a mapping table that can be used with the translate() method to replace specified characters.

Syntax

str.maketrans(x, y, z)

-  x -	Required. If only one parameter is specified, this has to be a dictionary describing how to perform the replace. If two or more parameters are specified, this parameter has to be a string specifying the characters you want to replace.
-  y -	Optional. A string with the same length as parameter x. Each character in the first parameter will be replaced with the corresponding character in this string.
-  z -	Optional. A string describing which characters to remove from the original string.

```python
txt = "Hello Sam!"

mytable = str.maketrans("S", "P")

print(txt.translate(mytable))
```

```python
txt = "Hi Sam!"
x = "mSa"
y = "eJo"
mytable = str.maketrans(x, y)
print(txt.translate(mytable))
```

Python Collections (Arrays)
There are four collection data types in the Python programming language:

- **List** is a collection which is **ordered and changeable**. **Allows duplicate** members.
- **Tuple** is a collection which is **ordered and unchangeable**. **Allows duplicate** members.
- **Set** is a collection which is **unordered, unchangeable, and unindexed**. **No duplicate** members.
- **Dictionary** is a collection which is **ordered and changeable**. **No duplicate** members.Python Collections (Arrays)


When we say a collection is ordered, it means:
- The elements have a defined sequence ‚Äî each item has a fixed position (index) in the collection.
- When you iterate over the collection, items come out in the same order they were inserted.
- You can reliably access elements by their position (like list[0], tuple[1]).


- Ordered = predictable sequence, index-based access possible.
- Unordered = no guaranteed sequence, you can‚Äôt rely on positions.

### Python frozenset

frozenset is an immutable version of a set.

Like sets, it contains unique, unordered, unchangeable elements.

Unlike sets, elements cannot be added or removed from a frozenset.

```python
# Creating a frozenset:

x = frozenset({"apple", "banana", "cherry"})
print(x)
print(type(x))
```

Being immutable means you cannot add or remove elements. However, frozensets support all non-mutating operations of sets.

- copy() - Returns a shallow copy.
- difference() - (-) Returns a new frozenset with the difference
- intersection() - (&) Returns a new frozenset with the intersection
- isdisjoint() - Returns whether two frozensets have an intersection
- issubset() - (<=/<) Returns True if this frozenset is a (proper) subset of another
- issuperset() - (>=/>) Returns True if this frozenset is a (proper) superset of another
- symmetric_difference() - (^) Returns a new frozenset with the symmetric differences
- union() - ( | ) Returns a new frozenset containing the union

### Online Solution

```python
def rank_students(students):
    # Sort by score descending
    students = sorted(students, key=lambda x: x[1], reverse=True)
    
    result = []
    prev_score = None
    prev_rank = 0
    
    for idx, (name, score) in enumerate(students, start=1):
        # If score is same as previous, use same rank
        if score == prev_score:
            rank = prev_rank
        else:
            rank = idx  # Olympic ranking: rank = position in sorted list
        
        result.append((name, score, rank))
        
        # Update trackers
        prev_score = score
        prev_rank = rank
    
    return result


students = [
    ("Alice", 95),
    ("Bob", 85),
    ("Charlie", 95),
    ("David", 80),
    ("Eve", 85),
    ("Frank", 70)
]

print(rank_students(students),end='')
```
